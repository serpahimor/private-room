```python
from typing import TypeVar, Generic, List                                                         
T = TypeVar('T')                                                                                   
class Stack(Generic[T]):                                                                           
    def __init__(self) -> None:                                                                           self.items = []                                                                           
    def push(self, item: T) -> None:                                                               
        self.items.append(item)                                                                   
    def pop(self) -> T:                                                                           
        return self.items.pop()                                                                   
st = Stack[int]()                                                                                 
st.push("A")                                                                                       
stack2 = Stack()                                                                                   
stack2.push("b")                                                                                   
stack2.push(2) 
```
